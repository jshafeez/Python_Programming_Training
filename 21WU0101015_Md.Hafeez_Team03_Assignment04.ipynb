
### 1. What exactly is []?
`[]` is an empty list in Python. It is a data structure used to store multiple items in a single variable. Lists are ordered, mutable (changeable), and can store items of any data type.

---

### 2. In a list of values stored in a variable called spam, how would you assign the value 'hello' as the third value? (Assume [2, 4, 6, 8, 10] are in spam.)
To assign 'hello' as the third value (index 2) in the list, you would write:
```python
spam[2] = 'hello'
```
This replaces the third value (which is 6) with 'hello'.

---

### 3. What is the value of spam[int(int('3' * 2) / 11)]?
Breaking it down:
- `'3' * 2` produces `'33'`.
- `int('33')` gives `33`.
- Dividing 33 by 11 gives `3`.
- `spam[3]` refers to the element at index 3 of the list `['a', 'b', 'c', 'd']`, which is `'d'`.

Thus, the value is **'d'**.

---

### 4. What is the value of spam[-1]?
`spam[-1]` refers to the last element of the list. In the list `['a', 'b', 'c', 'd']`, the last element is `'d'`.

Thus, the value is **'d'**.

---

### 5. What is the value of spam[:2]?
`spam[:2]` returns the slice of the list from the start to, but not including, index 2. In `['a', 'b', 'c', 'd']`, it returns `['a', 'b']`.

Thus, the value is **['a', 'b']**.

---

### 6. What is the value of bacon.index('cat')?
The `index()` method returns the index of the first occurrence of the specified value. In the list `[3.14, 'cat', 11, 'cat', True]`, the first occurrence of `'cat'` is at index 1.

Thus, the value is **1**.

---

### 7. How does bacon.append(99) change the look of the list value in bacon?
The `append()` method adds an item to the end of the list. After `bacon.append(99)`, the list becomes:
```python
[3.14, 'cat', 11, 'cat', True, 99]
```

---

### 8. How does bacon.remove('cat') change the look of the list in bacon?
The `remove()` method removes the first occurrence of the specified value from the list. After `bacon.remove('cat')`, the list becomes:
```python
[3.14, 11, 'cat', True, 99]
```
Note: Only the first `'cat'` is removed.

---

### 9. What are the list concatenation and list replication operators?
- **Concatenation operator (`+`)**: Joins two lists together.
  Example:
  ```python
  [1, 2] + [3, 4]  # Results in [1, 2, 3, 4]
  ```
- **Replication operator (`*`)**: Repeats the list a given number of times.
  Example:
  ```python
  [1, 2] * 3  # Results in [1, 2, 1, 2, 1, 2]
  ```

---

### 10. What is the difference between the list methods append() and insert()?
- **append()**: Adds an item to the end of the list.
- **insert()**: Adds an item at a specified index. It shifts the elements after the specified index to the right.

Example:
```python
spam.append(5)   
spam.insert(2, 5)  
```

---

### 11. What are the two methods for removing items from a list?
- **remove()**: Removes the first occurrence of the specified value.
- **pop()**: Removes the item at a specified index (or the last item if no index is provided).

---

### 12. Describe how list values and string values are identical.
Both lists and strings are ordered sequences. They support indexing, slicing, and iteration. However, lists can contain any type of objects (integers, strings, etc.), while strings specifically store sequences of characters.

---

### 13. What's the difference between tuples and lists?
- **Lists**: Mutable (can be changed after creation). They are created using square brackets `[]`.
- **Tuples**: Immutable (cannot be changed after creation). They are created using parentheses `()`.

---

### 14. How do you type a tuple value that only contains the integer 42?
To create a tuple with only the integer 42, you write:
```python
(42,)
```
The trailing comma is necessary to distinguish it from a regular parenthesis.

---

### 15. How do you get a list value's tuple form? How do you get a tuple value's list form?
- To convert a list to a tuple: `tuple(list_value)`
- To convert a tuple to a list: `list(tuple_value)`

---

### 16. Variables that "contain" list values are not necessarily lists themselves. Instead, what do they contain?
They contain **references** to the list objects in memory. This means changes made to the list through one variable will reflect in other variables referring to the same list.

---

### 17. How do you distinguish between copy.copy() and copy.deepcopy()?
- **copy.copy()**: Creates a **shallow copy** of the list, which means it copies the outer list but not any nested objects within it.
- **copy.deepcopy()**: Creates a **deep copy** of the list, which means it copies the entire list and all the nested objects within it.
